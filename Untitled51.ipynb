```python
import pandas as pd
from statsmodels.stats import weightstats as stests
     

df = pd.read_csv('/content/breast-cancer.csv')
df.head()
```

```python

df[['texture_mean','perimeter_mean']].describe()
     
```

```python
ztest ,pvalue = stests.ztest(df['texture_mean'], x2=None, value=146)
print(float(pvalue))
print(ztest)
     
```

```python
if pvalue<0.05:
    print("Null hyphothesis rejected , Alternative hyphothesis accepted")
else:
    print("Null hyphothesis accepted , Alternative hyphothesis rejected")
```

```python
ztest,pval1=stests.ztest(df['texture_mean'],df['perimeter_mean'],value=100)
print(float(pval1))
print(ztest)
```

```python
if pval1<0.05:
    print("Null hyphothesis rejected , Alternative hyphothesis accepted")
else:
    print("Null hyphothesis accepted , Alternative hyphothesis rejected")
```
