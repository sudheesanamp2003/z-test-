```python
import pandas as pd
from statsmodels.stats import weightstats as stests
```

```python
df = pd.read_csv('data.csv')
df.head()
```

```python
df[['texture_mean','radius_mean']].describe()
     
```

```python
ztest ,pvalue = stests.ztest(df['texture_mean'], x2=None, value=146)
print(float(pvalue))
print(ztest)
```

```python
if pvalue<0.05:
    print("Null hyphothesis rejected , Alternative hyphothesis accepted")
else:
    print("Null hyphothesis accepted , Alternative hyphothesis rejected")
    
```

```python
ztest,pval1=stests.ztest(df['texture_mean'],df['radius_mean'],value=100)
print(float(pval1))
```

```python
if pval1<0.05:
    print("Null hyphothesis rejected , Alternative hyphothesis accepted")
else:
    print("Null hyphothesis accepted , Alternative hyphothesis rejected")
     
```
